**Write A Program to ask user to enter distance from source to destination of his travelling , calculate fare with the rate of 12 rs per km, if fare is above 1000 give 10% discount and display final amount else display actual amount with message "no discount for short journeys"**

```python
dist = float(input('Enter distance '))
rate_per_km = 12
fare = dist * rate_per_km
final_amount = 0
if fare > 1000:
    final_amount = fare - fare * 0.1
    print(final_amount,'is Final amount')
else:
    print(fare,'No discount for short journeys')
```

**Write A Program to calculate grade of student, for this accept marks for 5 subjects, calculate total and percentage. 
Based on percentage display grade .. criteria is per>85 a+, >75 A, >60 "I',>50 "II", >40 "pass" else fail**

```python
# s1,s2,s3,s4,s5 = input('Enter marks for 5 subjects: ').split()
```

```python
s1,s2,s3,s4,s5 = [int(x) for x in input("Enter multiple value: ").split(",")]
```

```python
total = s1 + s2 + s3 + s4 + s5
print(total)
percentage = ((total * 100 ) / 500 )
if percentage >= 40:
    print('PASS')
elif percentage >= 50:
    print('II')
elif percentage >= 60:
    print('I')
elif percentage >=75:
    print('A')
elif percentage >= 85:
    print('A+')
else:
    print('FAIL')
```

# pyramid pattern printing

```python
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print('*',end=' ')       
        j = j+ 1
    print('\n')
    i = i + 1
# i is outer loop responsible for printing number of rows which are 5
# j is inner loop responsible for printing number of columns upto number of rows which is for 1st row, number of column is 1
```
